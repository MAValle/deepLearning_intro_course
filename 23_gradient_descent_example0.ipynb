# 2.3 Gradient descent understanding.
#### Example 0
 Tratemos de entender el proceso por el cual la red meuronal "aprende".

Supongamos una una función simple, con sólo 1 variable.  Para nuestro propósitos, equivale a una función de pérdida que deseamos minimizar. En otras palabras, deseamos encontrar el valor del parámetro que minimice la pérdida (o loss).

Suponga que esta función de loss sea:

$y = (x + 10)^{2}$


El **Gradient descent** o GD, es un método iterativo para optimizar una función objetivo, en nuestro caso. minimizar el loss.

En términos generales,

$ w:= w - \eta \nabla Q(w)$

donde $w$ es el conjunto de parámetros que deseamos encontrar para minimizar el loss, y $Q(w)$ es el loss. La constante $\eta$ es el learning rate.

En nuestro caso, el asunto se reduce a:

$ x := x - \eta \nabla y = x - 2\eta (x - 10)$


Entonces, partiendo de un valor inicial $x_0$, comenzamos a iterar.  
Por ejemplo hagamos $x_0=3$ y $\eta=0.01$:

$ x = 3 - 2*0.01*(3+10) = 2.74$

En la segunda iteración:

x = 2.74 - 2*0.01*(2.74+10) = 2.4852$

y así sucesivamente....

Veamos esto en Python.

```python
lr = 0.01
parada = 0.000001  # cuando la diferencia entre el x actual y el nuevo alcance este valor, paramos.
max_iter = 1000  # numero de iteraciones
iteracion = 0
x_val = 3
df = lambda x: 2*(x+10) # este es nuestro gradiente.
previous_step_size = 1
```

```python
# Programamos las iteraciones, de tal forma que cuando la diferencia de x entre dos
# iteraciones sea menor que "parada", o cuando el número de iteraciones
# alcance "epochs"


while previous_step_size > parada and iteracion < max_iter:
    prev_x = x_val # almacena el valor previo del parámetro
    x_val = x_val - lr * df(prev_x ) # GD
    previous_step_size = abs(x_val - prev_x) #Change in x
    iteracion = iteracion + 1 #iteration count
    print("Iteracion",iteracion,"\nX valor del parámetro es",x_val)

# note que el valor teórico deve ser igual a 10.
```
